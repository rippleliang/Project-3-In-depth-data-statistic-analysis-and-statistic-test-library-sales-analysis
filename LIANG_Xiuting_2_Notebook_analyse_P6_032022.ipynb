```python
import pandas as pd
import seaborn as sns
multipalette = sns.color_palette('muted')
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import numpy as np
import datetime as dt
import time
import calendar
from scipy import stats
from statsmodels.stats.anova import anova_lm
from statsmodels.formula.api import ols
import pylab
```

# L'analyse de vente

### Importer les données

```python
df = pd.read_csv('df_library.csv', index_col = 'index', dtype = {'categ':'int64'})
df['date'] = pd.to_datetime(df['date'])
df.head()
```

### Traiter les formats de datetime

```python
df['year'] = df['date'].dt.strftime("%Y")
df['month'] = df['date'].dt.strftime("%Y-%m")
df['short_date'] = df['date'].dt.strftime("%Y-%m-%d")
df['quarter'] = df['date'].dt.to_period('Q')
df['day'] = df['date'].dt.day_name()
df.head()
```

```python
df.info()
```

## Le chiffre d'affaires (CA) et la tendance

### CA total annuel

```python
# calculer le CA de chaque année
df_yCA = df.groupby (by = 'year').sum()['CA']
df_yCA = pd.DataFrame(df_yCA)
df_yCA['CA_million'] = round(df_yCA['CA']/1000000,2)
df_yCA
```

```python
# créer la graphique histogramme
%matplotlib inline
plt.figure(figsize=(12,6))
sns.barplot(x = df_yCA.index, y = df_yCA.CA_million, color = 'cornflowerblue')
plt.title("L’évolution du chiffre d'affaires annuel", fontsize = 20)
```

### L’évolution du CA par trimestre

```python
# calculer le CA de chaque trimestre
df_qCA = df.groupby (by = 'quarter').sum()['CA']
df_qCA = pd.DataFrame(df_qCA)
df_qCA['CA_million'] = round(df_qCA['CA']/1000000,2)
df_qCA
```

```python
# créer la graphique histogramme
%matplotlib inline
plt.figure(figsize=(12,6))
sns.barplot(x = df_qCA.index, y = df_qCA.CA_million, color = 'cornflowerblue')
plt.title("L’évolution du chiffre d'affaires par trimestre", fontsize = 20)
```

<font color = Indigo> ***Il n'y a pas d'impact à CA par saisonnalité.** </font>

### La tendance de CA par mois en moyenne mobile

```python
# calculer le CA de chaque mois
df_mCA = df.groupby (by = 'month').sum()['CA']
df_mCA = pd.DataFrame(df_mCA)
df_mCA['CA_mille'] = round(df_mCA['CA']/1000,2)
df_mCA
```

```python
# créer la graphique histogramme
%matplotlib inline
plt.figure(figsize=(12,6))
sns.lineplot(x = df_mCA.index, y = df_mCA['CA_mille'], color = 'cornflowerblue', label = 'CA_mille') # plot CA
sns.lineplot(x = df_mCA.index, y = df_mCA['CA_mille'].rolling(window =3, min_periods =1).mean(),  
             color = 'purple', 
             label = 'CA_moyenne_mobile') # plot CA en moyenne mobile
plt.title("La tendance de CA par mois en moyenne mobile", fontsize = 20)
plt.xticks(rotation = 60)
```

<font color = Indigo> ***Le CA a baissé en 2021-10** </font>

### La tendance de CA par jour en moyenne mobile

```python
# calculer le CA de chaque jour
df_dCA = df.groupby (by = 'short_date').sum()['CA']
df_dCA = pd.DataFrame(df_dCA)
df_dCA['CA_mille'] = round(df_dCA['CA']/1000,2)

# calculer le CA en moyenne mobile
df_dCA['CA_moyenne_mobile'] = df_dCA['CA_mille'].rolling(window =7, min_periods =1).mean()
df_dCA = df_dCA.reset_index()
df_dCA.head()
```

```python
# créer la graphique de courbe
x_dCA = df_dCA['short_date']
y_dCA1 = df_dCA['CA_mille']
y_dCA2 = df_dCA['CA_moyenne_mobile']
x_major_locator_dCA = MultipleLocator(90)

%matplotlib inline
fig, ax = plt.subplots(figsize=(15, 8), constrained_layout=True)
ax.plot (x_dCA, y_dCA1, color = 'cornflowerblue', label = 'CA_mille')
ax.plot (x_dCA, y_dCA2, color = 'purple', label = 'CA_moyenne_mobile')
ax.xaxis.set_major_locator(x_major_locator_dCA)
plt.title("La tendance de CA par jour en moyenne mobile", fontsize = 20)
ax.set_xlabel("jour")
ax.set_ylabel("CA")
ax.legend()
```

<font color = Indigo> ***Le CA a baissé en 2021-10** </font>

### Traiter la baisse en 2021-10

```python
# regrouper les donées par mois et catégorie
categ_month = df.groupby(['month', 'categ']).sum()[['CA']].unstack()

# créer la graphique
%matplotlib inline
categ_month.plot(
    figsize =(12,6), 
    ylabel = "Chiffre d'affaire",
    title = "chiffre d'affaire par catogory",
    colormap = 'tab10'
)
plt.title("L’évolution du chiffre par catégorie", fontsize = 20)
```

<font color = Indigo> ***La baisse de CA est à cause de la baisse de categ 1 en 2021-10.** </font>

#### Analyser la baisse de categ 1 en 2021-10

```python
# sélectionner tous les dates de 2021-10 et le CA pour categ1
df_10categ1 = df.loc[(df['month'] == '2021-10') & (df['categ'] == 1),:]
sns.barplot(x = df_10categ1['day'], y = df_10categ1['CA'],color = 'cornflowerblue')
plt.title("Le CA de categ 1 en 2021-10 ", fontsize = 20)
```

<font color = Indigo> ***Il n'y a que 5 jours en 2021-10 qui ont de données pour la vente de categ1** </font>

```python
# pour comparer la tendance de categ1 entre 2021-08 et 2021-11, sélectionner les données pour ces 4 mois
df_q3q4 = df.loc[df['month'].isin(['2021-08','2021-09','2021-10','2021-11']),:]

# sélectionner categ1
df_q3q4c1 = df_q3q4.loc[df_q3q4['categ'] == 1, :] 

# compter le nombre de vente chaque jour
df_c1 = df_q3q4c1.groupby (by = 'short_date' ).count()[['price']]
df_c1 = df_c1.rename(columns={'price':'nb_de_vente'})

# créer la graphique
plt.figure(figsize=(20,6))
sns.lineplot(x = df_c1.index, y = df_c1['nb_de_vente'], color = 'royalblue')
plt.xticks(rotation = 60, fontsize = 8)
plt.title("La tendance du nombre de vente categ 1 entre 2021-08 et 2021-11", fontsize = 20)
plt.show()
```

<font color = Indigo> ***La tendance de categ 1 est continu avec l'absence de données pendant 02-10-2021 et 27-10-2021** </font>

### La courbe de Lorenz des réfèrences

```python
# creér la courbe de Lorenz
df_prod_rank = df_prod_ca.sort_values(by = 'CA', ascending = False)
n_ref = len(df_prod_rank)
lorenz_ref = np.cumsum(np.sort(df_prod_rank['CA'])) / df_prod_rank['CA'].sum()
lorenz_ref = np.append([0],lorenz_ref) # La courbe de Lorenz commence à 0

plt.figure(figsize = (12,6))
xaxis = np.linspace(0-1/n_ref,1+1/n_ref,n_ref+1) #Il y a un segment de taille n pour chaque individu, plus 1 segment supplémentaire d'ordonnée 0. Le premier segment commence à 0-1/n, et le dernier termine à 1+1/n.
plt.plot(xaxis,lorenz_ref,drawstyle='steps-post', color = 'darkblue', linewidth = 5)
plt.xlabel('Pourcentage de produits accumulé')
plt.ylabel('Pourcentage de CA accumulé')
plt.plot ([0,0.93],[0.5, 0.5], c = 'crimson', linestyle='--')
plt.plot ([0.93, 0.93],[0,0.5], c = 'crimson', linestyle='--')
plt.plot ([0,0.79],[0.2, 0.2], c = 'purple', linestyle='dotted')
plt.plot ([0.79, 0.79],[0,0.2], c = 'purple', linestyle='dotted')
plt.plot ([0,1], c = 'cornflowerblue')
plt.text(0.75,0.55, 'position de produit \n avec CA médian \n(0.93, 0.5)',c = 'crimson',fontsize = 12)
plt.text(0.6,0.25, 'La loi de Pareto, 20/80',c= 'purple',fontsize = 12)
plt.title("La courbe de Lorenz des CA generés par les réfèrences", fontsize = 20)
plt.show()
```

```python
# calculer gini
AUC_ref = (lorenz_ref.sum() -lorenz_ref[-1]/2 -lorenz_ref[0]/2)/n_ref # Surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
S_ref = 0.5 - AUC_ref # surface entre la première bissectrice et le courbe de Lorenz
gini_ref = 2*S_ref
print ("gini: " + str(round(gini_ref,2)))
```

### Eta²

le rapport de corrélation, compris entre 0 et 1, donné par :

η2 Y/X=Vinterclasses/Vtotale

```python
# def eta_squared(x,y)
def eta_squared(x,y):
    moyenne_y = y.mean()
    classes = []
    for classe in x.unique():
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
    return SCE/SCT
```

```python
# calculer Eta² des ages et des catégories
eta2_age = eta_squared(df_client['age'],df_client['categ'])
print("Eta² = " + str(round(eta2_age,2)) + 
      "\n*Il y a une corrélation entre les catégories des livres achetés et les âges differents des clients.")
```

Cependant, il n'y a pas d'âge moins de 18 ans dans les profiles de clients, qui ne correspondent pas la distribution normale. Ici, on va ultiliser ANOVA quand même et le teste Kruskal Wallis dans le même temps comme le teste non-parametre.** </font>

### Teste ANOVA

<font color = Indigo> **Test ANOVA (ANalysis Of VAriance):** </font>

<font color = Indigo> **Question: Les catégories des livres achetés sont-ils toujour pareils entre les âges differents des clients ?** </font>
1. H0: Les catégories des livres achetés sont **pareils** entre les âges differents des clients.
2. HA: Les catégories des livres achetés sont **differents** entre les âges differents des clients.

#### Methode 1: statsmodels

```python
# calculer le p-value
model = ols('age ~ C(categ)', data = df_client).fit()
anova_lm(model)
```

```python
print ("*P-value < 0.05, on rejette H0 et accept HA: Les catégories des livres achetés sont differents entre les âges differents des clients. \nIl y a une corrélation entre les catégories des livres achetés et les âges differents des clients.")
```

#### Methode 2: scipy.stats
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.f_oneway.html

```python
# grouper les samples
categ0 = df_client[df_client['categ'] == 0]['age']
categ1 = df_client[df_client['categ'] == 1]['age']
categ2 = df_client[df_client['categ'] == 2]['age']

# calculer le p-value
F__categ, p_value_categ = stats.f_oneway (categ0, categ1, categ2)
print ("F-statistic = " + str(round(F__categ,2)))
print ("P-value = " + str(p_value_categ))
print ("*P-value < 0.05, on rejette H0 et accept " + 
       "HA: Les catégories des livres achetés sont differents entre les âges differents des clients. " +
       "\nIl y a une corrélation entre les catégories des livres achetés et les âges differents des clients.")
```

### Kruskal Wallis

```python
# Séparer les catégories
categ0 = df_client[df_client['categ'] == 0]['age']
categ1 = df_client[df_client['categ'] == 1]['age']
categ2 = df_client[df_client['categ'] == 2]['age']

# calculer le p-value
H_statistic, P_value_H = stats.kruskal(categ0, categ1, categ2, nan_policy='omit')
print ("Kruskal Wallis-statistic = " + str(round(H_statistic,2)))
print ("P-value = " + str(P_value_H))
print ("*P-value < 0.05, on rejette H0 et accept " + 
       "HA: Les catégories des livres achetés sont differents entre les âges differents des clients. " +
       "\nIl y a une corrélation entre les catégories des livres achetés et les âges differents des clients.")
```

### Kruskal-Wallis

<font color = Indigo> **Question : Les fréquences d’achats sont-ils toujour pareils entre les âges differents des clients ? (les âges vs. les fréquences des achats)** </font>
1. H0: Les fréquences d’achats sont toujour **pareils** entre les âges differents des clients. (Les âges et les fréquences des achats sont 2 variables **independents**.
